```python
def printBooksInfo(dic):
    print("\n"*2)
    print(f"Total {len(dic)} books:")
    print("-"*50)
    for book in dic:
        f = '{:^20}:{:<20}'
        copies=int(dic[book][3])+int(dic[book][4])
        print(f.format("serial#",book))
        print(f.format("title",dic[book][0]))
        print(f.format("number of authors",len(dic[book][1])))
        print(f.format("price",dic[book][2]))
        print(f.format("total copies",copies))
        print("-"*50)
    print()  
    
```

```python
def print_details(matchingBooksList,dic):
    print("Matched records:\n","*"*30)
    for book in matchingBooksList:
        f = '{:<20}:{:<20}'
        copies=int(dic[book][3])+int(dic[book][4])
        print(f.format("serial#",book))
        print(f.format("title",dic[book][0]))
        print("authors :")
        for auther in dic[book][1]:
            print(f"\t-{auther}")
        print(f.format("price:",dic[book][2]))
        print(f.format("copies in library:",dic[book][3]))
        print(f.format("borrowed copies",dic[book][4]))
        print("-"*50)
    
```

```python
def search_a_book(dic):
    c=False
    while not c:
        choice=input("Enter (t) to search by title or (a) to search by author name:")
        if choice.lower()=="a":
            choice="a"
            c=True
        elif choice.lower()=="t":
            choice="t"
            c=True
        else:
            print("invalied choice please try again")
    matchingBooksList=[]
    if choice=="t":
        title_part=input("Enter the title:")
        for i in dic:
            if title_part.lower() in dic[i][0].lower():
                matchingBooksList.append(i)
    if choice=="a":
        auther=input("Enter author name:")
        for i in dic:
            for auth in dic[i][1]:
                if auther.lower() in auth.lower():
                    matchingBooksList.append(i)
    if matchingBooksList==[]:
        print("No matched record found\n","-"*30)
        return
    print_details(matchingBooksList,dic)
```

```python
def add_a_new_book(dictOfItems):
    outfile=open("book.txt",'a')
    d={}
    list_of_errors=[]
    done=False
    serial=input("Enter serial number: ")
    book_title=input("Enter book title:")
    authers=[]
    count=1
    while True:
        auther=input(f"Enter name of author {count}:")
        if auther.lower() !="q":
            authers.append(auther)
        else:
            break
        count+=1
    price=input("Enter book price: ")
    copies=input("Enter number of book copies: ")
    if serial in dictOfItems:
        list_of_errors.append("Error: serial number is already used")
    if len(serial)>5:
        list_of_errors.append("Error: serial number should be 5 digits")
    if book_title =="":
        list_of_errors.append("Error: Title should not be empty")
    for i in authers:
        if i =="":
            list_of_errors.append("Error: names of all authors should not be empty")
    try:
        price=float(price)
    except:
        list_of_errors.append("Error: price should be a valid float number")
    try:
        copies=int(copies)
    except:
        list_of_errors.append("Error: number of copies should be a valid int number")
    for mess in list_of_errors:
        print(mess)
    if list_of_errors==[]:
        list_of_values=[book_title,authers,str(price),str(copies),str(0)]
        dictOfItems[serial]=list_of_values
        print("New book was added successfully")
```

```python
def remove_book(dictOfbooks):
    serial_number=input("Enter book serial number to remove: ")
    results=[]
    for book in dictOfbooks:
        if serial_number== book:
            results.append(serial_number)
    
    if results!=[]:
        print_details(results,dictOfbooks)
        sure_Q=input("Deleting the book .. Are you sure (y/n):")
        if sure_Q.lower()=="y":
            if dictOfbooks[serial_number][4]!= "0":
                print("Book cannot be removed: borrowed copies must be 0\n\n")
                return
            del dictOfbooks[serial_number]
            print("Book was removed successfully\n\n")
        else:
            print("operation is cancelled\n\n")
    else:
        print("there is not a book like that")
 
    
```

```python
def ListM():
    while True:
        print("""\n\nLibrary Management System
==============================
1. Print books info
2. Search a book
3. Add new book
4. Remove a book
5. Borrow a book 
6. Return a book
7. Exit
===============================""")
        choice=input('Enter your choice: ')
        if choice == "quit" or choice.lower()=="exit"  :
            return(0)
        elif choice.isnumeric():
            if int(choice)==1:
                return(1)
            elif int(choice)==2:
                return(2)
            elif int(choice)==3:
                return(3)
            elif int(choice)==4:
                return(4)
            elif int(choice)==5:
                return(5)
            elif int(choice)==6:
                return(6)
            elif int(choice)==7:
                return(0)
            else:
                print("this is not one of the choices, please choose one of the upove")
                continue
        else:
            print("unvalied value please try again")
            continue
```

```python
def borrow_book(booksinfo,borrowdbooks):
    serial_number= input("Enter book serial number to borrow: ")
    if len(serial_number)>5:
        print("Error: serial number should be 5 digits\n\n")
        return
    if serial_number not in booksinfo:
        print("Error: no matched serial number\n\n")
        return
    if booksinfo[serial_number][3]=="0":
        print("Error: no available copies in the library\n\n")
        return
    user_Id=input("Enter user ID: ")
    count=0
    for bow in borrowdbooks:
        if bow[1] ==user_Id:
            count+=1
    if count>2:
        print("Error: user cannot borrow more than 3 books\n\n")
        return
    bow=[serial_number,user_Id]
    for book in borrowdbooks:
        if book == bow:
            print("Error: user already borrowed a copy of the book\n\n")
            return
        
    booksinfo[serial_number][3]=str(int(booksinfo[serial_number][3])-1)
    booksinfo[serial_number][4]=str(int(booksinfo[serial_number][4])+1)
    borrowdbooks.append(bow)
    print("Book was borrowed successfully")
    
        
```

```python
def return_book(booksinfo,borrowdbooks):
    serial_number= input("Enter book serial number you borrowed: ")
    user_Id=input("Enter user ID: ")
    bow=[serial_number,user_Id]
    if bow in borrowdbooks:
        booksinfo[serial_number][3]=str(int(booksinfo[serial_number][3])+1)
        booksinfo[serial_number][4]=str(int(booksinfo[serial_number][4])-1)
        borrowdbooks.remove(bow)
        print("Book was returned successfully")
    else:
        print("Error: no matched record found in borrowedInfo.txt")
    
        
```

```python
def main():
    dictOfbooks = {}
    itemsFile = open("book.txt","r")
    for line in itemsFile:
        line=line.rstrip()
        infoList = line.split(",")
        key = infoList[0]
        value = infoList[1:len(infoList)]
        value[1]= value[1].split(":")
        dictOfbooks[key] = value
    borrowefile=open("borrowedInfo.txt","r")
    borrowdbooks=[]
    for line in borrowefile:
        line=line.rstrip()
        line=line.split(",")
        borrowdbooks.append(line)
    while True:
        ch=ListM()
        if ch==1:
            printBooksInfo(dictOfbooks)
        if ch==2:
            search_a_book(dictOfbooks)
        if ch==3:
            add_a_new_book(dictOfbooks)
        if ch==4:
            remove_book(dictOfbooks)
        if ch==5:
            borrow_book(dictOfbooks,borrowdbooks)
        if ch==6:
            return_book(dictOfbooks,borrowdbooks)
        if ch==0:
            itemsFile = open("book.txt","w")
            for i in dictOfbooks:
                dictOfbooks[i]=[dictOfbooks[i][0],
                                ":".join(dictOfbooks[i][1])\
                                ,str(dictOfbooks[i][2]),str(dictOfbooks[i][3])\
                                ,str(dictOfbooks[i][4])]
                itemsFile.write(f"{i},"+",".join(dictOfbooks[i])+"\n")
            borrowefile=open("borrowedInfo.txt","w")
            for i in borrowdbooks:
                borrowefile.write(",".join(i)+"\n")
            borrowefile.close()   
            itemsFile.close()
            return

# Run the program  
main()
```
